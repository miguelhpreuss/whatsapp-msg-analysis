```python
import matplotlib.pyplot as plt
import plotly.offline as py
import plotly.graph_objs as go
```

```python
data = []
```

```python
with open('arquivo.txt', 'r', encoding='utf-8') as f:
  for i, linha in enumerate(f):
    dia = linha[0:2]
    mes = linha[3:5]
    ano = linha[6:11]
    hora = linha[11:13]
    min = linha[14:16]

    try:
      int(dia)
      posicao = 0
      pegou_dois_pontos = 0
      while True:
        posicao = linha.find(":", posicao)
        if posicao == -1:
          break
        else:
          pegou_dois_pontos += 1
        posicao += 1
        if pegou_dois_pontos == 2:
          break

      nome = linha[19:posicao-1].split(" ")[0]
      palavras = linha[posicao:].split(" ")
      
      data.append({
        "dia" : dia,
        "mes" : mes,
        "ano" : ano,
        "hora" : hora,
        "min" : min,
        "nome" : nome,
        "palavras":palavras
      })
    except:
      pass
```

```python
#Mensagens por dia

msg_dia = {}
dias = []

for i in data:
    try:
        dia,mes = int(i["dia"]),int(i["mes"])
    except:
        continue
    d = "{}/{}".format(dia,mes)
    dias.append(d)
    nome = i["nome"]
    try:
        msg_dia[nome][d] += 1
    except:
        try:
            msg_dia[nome][d] = 1
        except:
            msg_dia[nome] = {
                d : 1
            }
```

```python
traces = list()

for nome in msg_dia:
    ydados = list(msg_dia[nome].values())
    xdados = list(msg_dia[nome].keys())
    trace = go.Scatter(x = xdados, y = ydados, mode = 'markers+lines', name = nome)
    traces.append(trace)
```

```python

import nbformat


# Criando Layout
layout = go.Layout(title='Mensagens por dia',
                   yaxis={'title':'Mensagens'},
                   xaxis={'title': 'Dia'})
# Criando figura que será exibida
fig = go.Figure(data=traces, layout=layout )
# Exibindo figura/gráfico
#
fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = list(set(dias))
    )
)

py.iplot(fig)
```
